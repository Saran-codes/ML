```python
train = emnist_train.to_numpy()
test = emnist_test.to_numpy()
```

```python
X_train = train[:,1:]
X_test = test[:,1:]
```

```python
Y_train = train[:,0]
Y_test = test[:,0]
```

```python
X_train=X_train/255
X_test=X_test/255
```

```python
Z_train= np.zeros((X_train.shape[0],len(np.unique(Y_train))))
Z_train.shape
```

```python
for i in range(X_train.shape[0]):
    Z_train[i][Y_train[i]-1] = 1

Z_train
```

```python
def function(X,W,b):
    z=np.dot(X,W)+b
    f = 1/(1+np.exp(-z))
    return f

def err(X,Y,W,b):
    err = function(X,W,b)-Y
    return err

def loss_function(X,Y,W,b):
    m=X.shape[0]
    f1 = np.log(function(X,W,b))
    f2 = np.log(1-function(X,W,b))
    loss = -np.sum(np.multiply(Y,f1)+np.multiply((1-Y),f2))
    loss = loss/m
    return loss

def gradient_compute(X,Y,W,b):
    m=X.shape[0]
    dj_dw = np.dot(X.T,err(X,Y,W,b))
    dj_dw = dj_dw/m
    
    dj_db = np.sum(err(X,Y,W,b),axis =0)/m
    
    return dj_dw,dj_db

def gradient_descent(X,Y,W,b,L,n):
    m=X.shape[0]
    
    for i in range(n):
        dj_dw = gradient_compute(X,Y,W,b)[0]
        dj_db = gradient_compute(X,Y,W,b)[1]
        W=W-L*dj_dw
        b=b-L*dj_db
        loss = loss_function(X,Y,W,b)
        print(i, loss)
        
    return W,b,loss
```

```python
w_in = np.zeros((X_train.shape[1],Z_train.shape[1]))
b=0
```

```python
w_final, b_final,loss =gradient_descent(X_train,Z_train,w_in,b,0.96,500)
```

```python
Y_output = function(X_test,w_final,b_final)
Y_output
```

```python
max = Y_output.max(axis=1)
for i in range(X_test.shape[0]):
    for j in range(len(np.unique(Y_train))):
        if Y_output[i][j] == max[i]:
            Y_output[i][j] = 1
        else:
            Y_output[i][j] = 0
            
Y_output
```

```python
Y_predicted = np.empty_like(Y_test)
for i in range(X_test.shape[0]):
    for j in range(len(np.unique(Y_train))):
        if Y_output[i][j]==1:
            Y_predicted[i]=j+1
            
Y_predicted
```

```python
def accuracy(yout,ytest):
    m = yout.shape[0]
    count=0
    for i in range(m):
        if yout[i]==ytest[i]:
            count+=1
            
    acc = count/m*100
    return acc
```
