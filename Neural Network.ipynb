```python
def function(X,W,b):
    a = 1/(1+np.exp(-(np.dot(W,X)+b)))
    return a

def forward_prop(X,W1,b1,W2,b2):
    A1 = function(X,W1,b1)
    A2  = function(A1,W2,b2)
    return A1,A2
```

```python
def gradient_descent(X,Y,W1,b1,W2,b2,L,itr):
    for i in range(itr):
        m = X.shape[1]
        
        #forward
        A1 = 1/(1+np.exp(-(np.dot(W1,X)+b1)))
        A2 = 1/(1+np.exp(-(np.dot(W2,A1)+b2)))
        
        #loss
        f1 = np.log(A2)
        f2 = np.log(1-A2)
        loss = -np.sum(np.multiply(Y,f1)+np.multiply((1-Y),f2))/m
        
        #backward
        dj_dw2 = np.dot((A2-Y)*(A2*(1-A2)),A1.T)/m
        dj_db2 = np.dot((A2-Y)*(A2*(1-A2)),np.ones((m,1)))/m
        dj_dw1 = np.dot(np.dot(W2.T,(A2-Y)*(A2*(1-A2)))*(A1*(1-A1)),X_train.T)/m
        dj_db1 = np.dot(np.dot(W2.T,(A2-Y)*(A2*(1-A2)))*(A1*(1-A1)),np.ones((m,1)))/m
        
        #updating
        W2 = W2 - L*dj_dw2
        b2 = b2 - L*dj_db2
        W1 = W1 - L*dj_dw1
        b1 = b1 - L*dj_db1
        
        print(i,loss)
    return W1,b1,W2,b2
```

```python
def accuracy(yout,ytest):
    m = yout.shape[0]
    count=0
    for i in range(m):
        if yout[i]==ytest[i]:
            count+=1
            
    acc = count/m*100
    return acc
```
